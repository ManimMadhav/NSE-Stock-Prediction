```python
import os
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

from sklearn.preprocessing import MinMaxScaler
from sklearn.model_selection import train_test_split

from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense
from tensorflow.keras.layers import LSTM
```

```python
#using around 7 months of NIFTY data
df = pd.read_csv("nifty-janTojul.csv")
print(df.head())
```

```python
#data preprocessing
df.drop("Adj Close", inplace=True, axis=1)
print(df.head())
```

```python
#feature scaling
scaler = MinMaxScaler(feature_range=(0,1))
df[['Open','High','Low','Close','Volume']] =scaler.fit_transform(df[['Open','High','Low','Close','Volume']])
print(df.head())
```

```python
#split dataset into training set and test set
x = df.loc[:,['Open','High','Low','Volume']]
y = df.loc[:,['Close']]
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(x,y,test_size=0.2,random_state=0)

#reshape xtrain & xtest in 3-D
x_train =x_train.values.reshape(x_train.values.shape[0],x_train.values.shape[1] , 1)
x_test = x_test.values.reshape(x_test.values.shape[0],x_test.values.shape[1] , 1)
```

```python
#LSTM

model=Sequential()
model.add(LSTM(50,return_sequences=True,input_shape=(4,1)))
model.add(LSTM(50,return_sequences=True))
model.add(LSTM(50))
model.add(Dense(5,input_shape=(97,5)))
model.compile(loss='mean_squared_error',optimizer='adam')

train_predict=model.predict(x_train)
test_predict=model.predict(x_test)
train_predict=scaler.inverse_transform(train_predict)
test_predict=scaler.inverse_transform(test_predict)
```
